```python
frutas={"Platano":1.35,
             "Manzana":0.80,
             "Pera":0.85,
             "Naranja":0.70}


def getFruta():
    
    print("Frutas disponibles: ")
    for i in frutas:
        print(i)

    try:
        precio = frutas.get(input("INTRODUCE LA FRUTA QUE DESEAS SOLICITAR:" ))
        print(str(precio)+"€")
        opcion = input("¿quieres comprar? (1)si (x)no")
        if opcion == '1':
            
            
            cantidad = int(input("Cuantos kg kieres:  "))
            print("Gracias por comprar "+str(cantidad)+"kg "+ "por un total de "+str(cantidad*precio)+"€")
        
    except: print("\n fruta introducida no existe \n")
    
    
def añadirFruta():
    
    nombre:str=str(input("Introduce el nombre de la nueva fruta para añdadir:"))

    if  nombre.isalpha()==False:
        print("error de fruta")
        return;
    else:
        print()
    
    try:
    
    
        precio = float(input("introduce la cantidad en € por Kg: "))
        if precio <=0:
            print("error de precio")
            return;
    except:
        print("error de precio")
        return;
    
    frutas[nombre]=precio


def salir():
    print("GRacias")
    exit()



while True:
    
    print("Menú:")
    print("1. Consultar el precio de una fruta")
    print("2. Añadir una nueva fruta y su precio")
    print("3. Salir del programa")

    opcion = input("Selecciona una opción: ")

    if opcion == '1':
        
        getFruta();

             
    elif opcion == '2':
        
        añadirFruta();



    elif opcion == '3':
        
        salir()

    else:
        print("Opción no válida. Por favor, elige una opción del menú.")
```
